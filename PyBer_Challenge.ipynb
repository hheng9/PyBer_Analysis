### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
total_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_count.head()
```

```python
# 2. Get the total drivers for each city type
sum_by_drivers = pyber_data_df.groupby(["type"]).sum()["driver_count"]
sum_by_drivers
```

```python
#  3. Get the total amount of fares for each city type
sum_by_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
sum_by_fares
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride = sum_by_fares / total_ride_count
avg_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_drivers = sum_by_fares / sum_by_drivers
avg_fare_drivers
```

```python
#  6. Create a PyBer summary DataFrame. 
# summary_df = pd.DataFrame()

# summary_df["Total Rides"] = total_ride_count
# summary_df["Total Drivers"] = sum_by_drivers
# summary_df["Total Fares"] = sum_by_fares
# summary_df["Average Fare per Ride"] = avg_fare_ride
# summary_df["Average Fare per Driver"] = avg_fare_drivers

# summary_df

summary_df = pd.DataFrame({
    "Total Rides":total_ride_count,
    "Total Drivers":sum_by_drivers,
    "Total Fares":sum_by_fares,
    "Average Fare per Ride":avg_fare_ride,
    "Average Fare per Driver":avg_fare_drivers
})

summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
summary_df["Total Rides"] = summary_df["Total Rides"].map('{:,}'.format)
summary_df["Total Drivers"] = summary_df["Total Drivers"].map('{:,}'.format)
summary_df["Total Fares"] = summary_df["Total Fares"].map('${:,.2f}'.format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map('${:,.2f}'.format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map('${:,.2f}'.format)
summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
group_sum_fare = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
group_sum_fare

# group_sum_fare = pyber_data_df.groupby(["type","date"]).sum()["fare"]
# group_sum_fare
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
group_sum_fare = group_sum_fare.reset_index()
group_sum_fare.head()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_sum_fare = group_sum_fare.pivot(index='date', columns='type', values='fare')
pivot_sum_fare.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
specific_fare_dates = pivot_sum_fare.loc['2019-01-01':'2019-04-28']
specific_fare_dates.head(10)
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
specific_fare_dates.index = pd.to_datetime(specific_fare_dates.index)
specific_fare_dates.head(10)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
specific_fare_dates.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
specific_fare_weeks = specific_fare_dates.resample('W').sum()
specific_fare_weeks.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

specific_fare_weeks.plot(figsize=(15, 5))
plt.title("Total Fare by City Type", fontsize=18)
plt.ylabel("Fare ($USD)", fontsize=15)
plt.xlabel("Month", fontsize=15)
plt.grid(True)
plt.legend(["Rural","Suburban","Urban"], loc='center')
plt.show()
plt.savefig("analysis/PyBer_fare_summary.png")
```
